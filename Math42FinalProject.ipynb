```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("rideshare_kaggle.csv")
data = data.dropna()
data.shape
```

```python
plt.hist(data["price"], density=True)
plt.xlabel("Price")
plt.ylabel("Proportion of Sample")
plt.title("Frequency of Price for Ride Share Service")
```

```python
plt.scatter(data["distance"], data["price"])
plt.xlabel("Distance of Ride")
plt.ylabel("Price of Ride")
plt.title("Price of Ride vs. Distance of Trip")
```

```python
plt.scatter(data["surge_multiplier"], data["price"])
plt.xlabel("Surge Multiplier")
plt.ylabel("Price of Ride")
plt.title("Price of Ride vs. Surge Multiplier")
```
